```
#Movies Dataset
#Author: Pranav Thaenraj
#Date: 06/10/20

#Dataset from https://www.kaggle.com/rounakbanik/the-movies-dataset?select=movies_metadata.csv

# load both the csv file from the path 
#credit_file_location = "dbfs:/FileStore/tables/credits_small.csv"
movie_metadata_file_location = "dbfs:/FileStore/tables/movies_metadata_csv.001"
# call the read method pass the csv value for format 
infer_schema = "false"
first_row_is_harder = "true"
delimiter = ","
movie_metadata_df = ( spark.read.format("csv").option("header","true") 
  .option("header",first_row_is_harder) 
  .option("sep",delimiter).load(movie_metadata_file_location) )
display(movie_metadata_df.take(10))
```

```
#Exploding Movies MetaData Df
from pyspark.sql.functions import from_json,expr
from pyspark.sql.types import StructType, StructField, StringType, ArrayType

#Creating Struct for Belong_to_collection
belongs_to_collection_schema = StructType(
    [
        StructField('id', StringType(), True),
        StructField('name', StringType(), True),
        StructField('poster_path', StringType(), True)
    ]
)
#Exploding the belongs_to_collection collumn, selecting wanted columns
movie_metadata_json_df = (movie_metadata_df.withColumn ("belongs_to_collection", from_json ("belongs_to_collection", belongs_to_collection_schema))).select("adult","belongs_to_collection.*","budget","genres",expr("id as movie_ID"),"original_title","popularity","production_companies")


display(movie_metadata_json_df)


```

```
#cleansing data
movie_metadata_json_df = movie_metadata_json_df.na.drop()
#display(movie_metadata_json_df)
#movie_metadata_json_df.printSchema()



#Exploding genres column
#Creating Schema 
genres_schema = ArrayType(StructType(
    [
        StructField('id', StringType(), True),
        StructField('name', StringType(), True),
    ]
))
movie_metadata_json_df = (movie_metadata_json_df.withColumn ("genres", from_json ("genres", genres_schema))).select ("adult" ,expr("id as colection_id"),expr("name as collection_name"),"budget","movie_ID","original_title","popularity","production_companies","genres.id", "genres.name" )

#using the actual explode function to split the ArrayType into genre key value pairs
from pyspark.sql.functions import explode
movie_metadata_json_df = movie_metadata_json_df.withColumn("genre", explode(movie_metadata_json_df['name'])).select("adult","colection_id","collection_name","genre","budget","movie_ID","original_title","popularity","production_companies")

display(movie_metadata_json_df)

```

```
#The genres of each movie collection have been sepreated, which movie collection has the most genres
movie_grouped_df = movie_metadata_json_df.groupBy("collection_name").count().orderBy("count")
display(movie_grouped_df)


#From this we can see that the James bond collection and the Godzilla(Showa) collection have the most genres - at 70 and 32 respectively
```

```
#Filtering low-budget films 

movie_metadata_json_df = movie_metadata_json_df.withColumn("budget", movie_metadata_json_df["budget"].cast("Integer"))
movie_metadata_json_df = movie_metadata_json_df.where(movie_metadata_json_df["budget"]> 200000)

#Filtering Adult Content
Children_df = movie_metadata_json_df.where(movie_metadata_json_df['adult'] == False)
display(Children_df)
```

```
#The data that has been filtered out now is all technically considered to be chilren's content, so from here we will look to filter out 10 of the best chilrens films 
#filtering The best genres for chilren by genre
Children_df_filtered = Children_df.where(Children_df["genre"] == "Animation")
#all of the animation movies that are liste here are apt for a child audience, but what are the most popular of these movies?

Children_df_filtered = Children_df_filtered.withColumn("popularity", Children_df_filtered.popularity.cast("Integer")).na.drop().orderBy(Children_df_filtered["popularity"].desc()).take(10)
display(Children_df_filtered)
```

```
#Two Production studios that are known to produce kid-friendly content are Walt Disney and Universal, These two production studios are known to provide kid friendly content

#Exploding the production studio column
#Creating struct
from pyspark.sql.types import StructType, StructField, StringType,ArrayType,MapType
productioncompany_schema = StructType(
  [
     StructField('Studio', StringType(), True),
     StructField('Studio_id', StringType(), True),
  ]

)

Children_df = Children_df.withColumn ("production_companies", from_json ("production_companies", productioncompany_schema)).select("colection_id","collection_name","budget","movie_ID","original_title","popularity","production_companies.*")

#This Graph Shows the average budget of the first 10 collection of films in the movie_metadata_json_df 
display(movie_metadata_json_df.select("budget","collection_name").groupBy("collection_name").avg("budget").toDF("collection_name","avg(budget)").take(10))

```

```
#This graph Shows the correlation between budget and popularity for the first 10 films in movie_metadata_json_df
display(movie_metadata_json_df.select("budget","popularity").take(30))

#The scatterPlot below shows that there is not necessarily a positive linear correlation between popularity and budget
```

```
# explode json  coloums 
from pyspark.sql import functions as F
from pyspark.sql.functions import explode
from pyspark.sql.functions import from_json, col
from pyspark.sql.functions  import split 
from pyspark.sql.types import StructType, StructField, StringType
from pyspark.sql.types import  ArrayType 
#ref :https://stackoverflow.com/questions/51070251/pyspark-explode-json-in-column-to-multiple-columns 
#     https://datanoon.com/blog/semi_structured_data_spark/
belongs_to_collection_schema = StructType(
    [
        StructField('id', StringType(), True),
        StructField('name', StringType(), True),
        StructField('poster_path', StringType(), True)
    ]
)
movie_metadata_json_df = (
       movie_metadata_df
       .withColumn ("belongs_to_collection", from_json ("belongs_to_collection", belongs_to_collection_schema) )
        .select (  col ("genres"),  col ("adult")  , col ("belongs_to_collection.*") ,col("budget") ) 
                        ) 

movie_metadata_json_df = (
      movie_metadata_df
           .withColumn ("belongs_to_collection", from_json ("belongs_to_collection", belongs_to_collection_schema) )
           .select (  col ("genres"),  col ("adult")  , col ("belongs_to_collection.*")   ,col("budget"), 
			movie_metadata_df.id,
            movie_metadata_df.imdb_id	, movie_metadata_df.original_language,
			movie_metadata_df.original_title, movie_metadata_df.overview,
			movie_metadata_df.popularity,	movie_metadata_df.poster_path,
			movie_metadata_df.production_companies,	
			movie_metadata_df.production_countries,	
			movie_metadata_df.release_date,	
			movie_metadata_df.revenue,	
			movie_metadata_df.runtime	,
			movie_metadata_df.spoken_languages,	
			movie_metadata_df.status,	
			movie_metadata_df.tagline,	
			movie_metadata_df.title,	
			movie_metadata_df.video,	movie_metadata_df.vote_average,	movie_metadata_df.vote_count	
                ) 
           ) 

print (type(movie_metadata_json_df))
movie_metadata_json_df.printSchema() 
display(movie_metadata_json_df.show(1)) 
```

```
# create schema to hold the genres >>> ArrayType(StructType) 
productioncompany_schema = ArrayType(
          StructType([StructField('name', StringType(), nullable=True),
          StructField('id', StringType(), nullable=True) ]  )
                                        )
movie_data_genres_df = ( 
   movie_metadata_json_df
  .withColumn ("genres", from_json ("genres", genres_schema) )
  .withColumn ("production_companies", from_json ("production_companies", productioncompany_schema) )
  .select (col ("budget"),col("genres.id").alias ("genres_id" )
  ,col("genres.name").alias("genres_name")  ,
   col ("production_companies.name").alias ("production_companies_name"  ),
   col ("production_companies.id").alias ("production_companies_id"  ) ,
   movie_metadata_json_df.imdb_id,
   movie_metadata_df.original_language,
   movie_metadata_json_df.original_title, 
   movie_metadata_df.overview,
   movie_metadata_json_df.popularity,
   movie_metadata_df.poster_path,
   movie_metadata_json_df.release_date,
   movie_metadata_json_df.revenue,	
   movie_metadata_json_df.runtime	,
   movie_metadata_json_df.spoken_languages,	
   movie_metadata_json_df.status,	
   movie_metadata_json_df.tagline,	
   movie_metadata_json_df.title,	
   movie_metadata_json_df.video,	
   movie_metadata_json_df.vote_average
      )
)

display (movie_data_genres_df.show(10))
#movie_data_genres_df.collect()
# keep the result as parquet file 
movie_data_genres_df.write.parquet('dbfs:/FileStore/movie_data_geres')
```
